```python
import timeit

import matplotlib.pyplot as plt

from list_ import UnorderedList
from linear_data_structure import Stack, Queue
from linear_data_structure_ll import Stack as LinkedStack
from linear_data_structure_ll import Queue as LinkedQueue


plt.style.use('seaborn')

sizes = range(100, 10001, 200)
number = 1000


def plot_time_complexity(sizes, times, xlabel, ylabel, labels):
    
    fig, ax = plt.subplots()
    
    ax.plot(sizes, times[0], marker='.', linestyle='none', label=labels[0])
    ax.plot(sizes, times[1], marker='^', linestyle='none', label=labels[1])
    
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    plt.legend()
    plt.show()
```

#### List

```python
t = timeit.Timer('ls.append(0)', 'from __main__ import ls')

ls_times = []
ll_times = []

for s in sizes:
    
    ls = list(range(s))
    ls_times.append(t.timeit(number=number))
    
    ls = UnorderedList()
    for x in range(s):
        ls.append(x)
    ll_times.append(t.timeit(number=number))
```

```python
plot_time_complexity(sizes, (ls_times, ll_times),
                     'Size of list',
                     'Time to complete append operation (milliseconds)',
                     ['Python list', 'Linked list'])
```

#### Stack

```python
t = timeit.Timer('stack.push(0)', 'from __main__ import stack')

s_ls_times = []
s_ll_times = []

for s in sizes:
    
    stack = Stack()
    for x in range(s):
        stack.push(x)
    s_ls_times.append(t.timeit(number=number))
    
    stack = LinkedStack()
    for x in range(s):
        stack.push(x)
    s_ll_times.append(t.timeit(number=number))
```

```python
plot_time_complexity(sizes, (s_ls_times, s_ll_times),
                     'Size of stack',
                     'Time to complete push operation (milliseconds)',
                     ['Python list based stack', 'Linked list based stack'])
```

#### Queue

```python
t = timeit.Timer('q.enqueue(0)', 'from __main__ import q')

q_ls_times = []
q_ll_times = []

for s in sizes:
    
    q = Queue()
    for x in range(s):
        q.enqueue(x)
    q_ls_times.append(t.timeit(number=number))
    
    q = LinkedQueue()
    for x in range(s):
        q.enqueue(x)
    q_ll_times.append(t.timeit(number=number))
```

```python
plot_time_complexity(sizes, (q_ls_times, q_ll_times),
                     'Size of queue',
                     'Time to complete enqueue operation (milliseconds)',
                     ['Python list based queue', 'Linked list based queue'])
```
